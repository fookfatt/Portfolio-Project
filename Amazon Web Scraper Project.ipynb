```python
#import libraries
from bs4 import BeautifulSoup
import requests
import time
import datetime
import pandas as pd
import csv
import smtplib
```

```python
#url of item interested
url = 'https://www.amazon.sg/Sony-WH1000XM3B-Cancelling-Headphones-manufacturer/dp/B07H2DBFQZ/ref=sr_1_2?dchild=1&keywords=SONY+WH-1000XM3&qid=1630051967&sr=8-2'

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:91.0) Gecko/20100101 Firefox/91.0", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

page = requests.get(url, headers=headers)

#pulling the html data
soup1 = BeautifulSoup(page.content, 'html.parser')

#prettify() helps rearrange the text for easier accesibility 
soup2 = BeautifulSoup(soup1.prettify(),'html.parser')

title = soup2.find(id="productTitle").get_text()
price = soup2.find(id="priceblock_ourprice").get_text()
rating = soup2.find(class_="a-icon-alt").get_text()
```

```python
price = price.strip()[2:8]
title = title.strip()[0:52]
rating = rating.strip()
date_time = datetime.datetime.today()
```

```python
header = ['Title','Price','Rating','Date_time']
data = [title,price,rating,date_time]

#create csv,'w' - write
with open('Amazon_Web_Scraper_Dataset.csv','w',newline='',encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
```
